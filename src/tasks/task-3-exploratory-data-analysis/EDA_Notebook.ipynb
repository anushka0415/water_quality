# Explanatory Data Analysis for Kaliyasot Dam

### Importing necessary libraries

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

### Loading the Dataframe

```python
df1 = pd.read_csv("2022.csv")
df2 = pd.read_csv("2021.csv")
df = pd.concat([df1, df2])
# removing null values
df = df.dropna()
# setting temperature to celcius
df["wst"]+=273
# seeing some values
df
```

## EDA

```python
df.describe()
```

### Histograms for each variable

```python
fig = px.histogram(df, y=["ndwi", "mndwi", "ndci", "ndti", "do", "ndsi", "ph", "chl_a", "ssc", "wst"], x=df["date"])
fig.show()
```

### Heatmap of the correlation between variables

```python
corr = df.corr()
fig = go.Figure(data=go.Heatmap(z=corr.values, x=corr.index.values, y=corr.columns.values))
fig.show()
```

### Scatter Plots

```python
fig = px.scatter_matrix(df, dimensions=["ndwi", "mndwi", "ndci", "ndti", "do"], color='date')
fig.show()
```

```python
fig = px.scatter_matrix(df, dimensions=["ndsi", "ph", "chl_a", "ssc", "wst"], color='date')
fig.show()
```

### Line graphs of variables over time

```python
fig = px.line(df, x='date', y=df.columns)
fig.show()
```

### Scatter plots of each variable against a specific variable

```python
fig = px.scatter(df, x='date', y=df.columns)
fig.show()
```

### Plots for each variable

```python
variables = ["ndwi", "mndwi", "ndci", "ndti", "do", "ndsi", "ph", "chl_a", "ssc", "wst"]
for var in variables:
    fig = px.histogram(df, y=var, x=df["date"])
    fig.show()
```
