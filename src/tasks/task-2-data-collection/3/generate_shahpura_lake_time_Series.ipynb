```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Get resources from Drive
ndci_df = pd.read_csv("Chlorophyll.csv")
ndci_df = ndci_df.rename(columns={'Chlorophyll': 'chlorophyll'})
ndci_df = ndci_df.rename(columns={'Unnamed: 0': 'date'})
ndci_df.head()
```

```python
# Get resources from Drive
do_df = pd.read_csv("Dissolved Oxygen.csv")
do_df = do_df.rename(columns={'Dissolved Oxygen': 'dissolvedOxygen'})
do_df = do_df.rename(columns={'Unnamed: 0': 'date'})

do_df.head()
```

```python
# Get resources from Drive
ph_df = pd.read_csv("pH.csv")
ph_df = ph_df.rename(columns={'pH': 'ph'})
ph_df = ph_df.rename(columns={'Unnamed: 0': 'date'})

ph_df.head()
```

```python
# Get resources from Drive
salinity_df = pd.read_csv("Salinty.csv")
salinity_df = salinity_df.rename(columns={'Salinty': 'salinity'})
salinity_df = salinity_df.rename(columns={'Unnamed: 0': 'date'})
salinity_df.head()
```

```python
turbidity_df = pd.read_csv("Turbidity.csv")
turbidity_df = turbidity_df.rename(columns={'Turbidity': 'turbidity'})
turbidity_df = turbidity_df.rename(columns={'Unnamed: 0': 'date'})
turbidity_df.head()
```

```python
complete_df = pd.concat([pd.DataFrame(ndci_df, columns= ['date', 'chlorophyll']),\
                        pd.DataFrame(turbidity_df, columns=['turbidity']),\
                        pd.DataFrame(salinity_df, columns=['salinity']), \
                         pd.DataFrame(ph_df, columns=['ph']), \
                         pd.DataFrame(do_df, columns=['dissolvedOxygen'])], axis=1, sort=False)

complete_df.head()
```

```python
complete_df.to_csv("bhopal_shahpura_lake_time_series.csv")
```
