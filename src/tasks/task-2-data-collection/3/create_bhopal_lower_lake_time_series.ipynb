```python
import pandas as pd
import numpy as np
```

```python
# Get resources from Drive
ndci_df = pd.read_csv("/content/drive/MyDrive/NDCI_lower_lake_time_series/NDCI_time_series_lower_lake.csv")
ndci_df = ndci_df.rename(columns={'value': 'chlorophyll'})
ndci_df.head()
```

```python
ndsi_df = pd.read_csv("/content/drive/MyDrive/NDSI_lower_lake_time_series/NDSI_time_series_lower_lake.csv")
ndsi_df = ndsi_df.rename(columns={'value': 'salinity'})
ndsi_df.head()
```

```python
ndti_df = pd.read_csv("/content/drive/MyDrive/NDTI_lower_lake_time_series/NDTI_time_series_lower_lake.csv")
ndti_df = ndti_df.rename(columns={'value': 'turbidity'})
ndti_df.head()
```

```python
ph_df = pd.read_csv("/content/drive/MyDrive/PH_lower_lake_time_series/PH_time_series_lower_lake.csv")
ph_df = ph_df.rename(columns={'value': 'ph'})
ph_df.head()
```

```python
oxygen_df = pd.read_csv("/content/drive/MyDrive/dissolvedOxygen_lower_lake_time_series/dissolvedOxygen_time_series_lower_lake.csv")
oxygen_df = oxygen_df.rename(columns={'value': 'dissolvedOxygen'})
oxygen_df.head()
```

```python
complete_df = pd.concat([pd.DataFrame(ndci_df, columns= ['date', 'chlorophyll']),\
                        pd.DataFrame(ndti_df, columns=['turbidity']),\
                        pd.DataFrame(ndsi_df, columns=['salinity']), \
                         pd.DataFrame(ph_df, columns=['ph']), \
                         pd.DataFrame(oxygen_df, columns=['dissolvedOxygen'])], axis=1, sort=False)

complete_df.head()
```

```python
complete_df.to_csv("/content/drive/MyDrive/Water_Quality/bhopal_lower_lake_time_series.csv")
```
