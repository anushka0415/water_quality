```python
# define date range
start_date = '2021-01-01'
end_date = '2022-12-31'
```

```python
# define bands to extract
bands_ndwi = ['B3', 'B8']
bands_ndci = ['B4', 'B3']
bands_do = ['B4', 'B5']
bands_temp = ['B10']

# create Sentinel-2 image collection
sentinel = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
                .filterBounds(region) \
                .filterDate(start_date, end_date) \
                .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20)) \
                .median()

#numpy data variables
ndwi = sentinel.normalizedDifference(bands_ndwi).rename('ndwi')
ndci = sentinel.normalizedDifference(bands_ndci).rename('ndci')
dissolved_oxygen = sentinel.normalizedDifference(bands_do).rename('dissolvedoxygen')
ndti = sentinel.normalizedDifference(['B8A', 'B12']).rename('ndti')
ndsi = sentinel.normalizedDifference(['B3', 'B11']).rename('ndsi')

#nwdi data
latlon = ee.Image.pixelLonLat().addBands(ndwi)
latlon = latlon.reduceRegion(
    reducer=ee.Reducer.toList(),
    geometry=region,
    scale=100)
data_ndwi = np.array((ee.Array(latlon.get('ndwi')).getInfo()))

#ndci data
latlon = ee.Image.pixelLonLat().addBands(ndci)
latlon = latlon.reduceRegion(
    reducer=ee.Reducer.toList(),
    geometry=region,
    scale=100)
assert len(bands_ndci) == 2, 'bands_ndci should contain exactly 2 band names'
data_ndci = np.array((ee.Array(latlon.get('ndci')).getInfo()))

#dissolved oxygen data
latlon = ee.Image.pixelLonLat().addBands(dissolved_oxygen)
latlon = latlon.reduceRegion(
    reducer=ee.Reducer.toList(),
    geometry=region,
    scale=100,
    tileScale=16)
data_oxygen = np.array((ee.Array(latlon.get('dissolvedoxygen')).getInfo()))


#ndti data
latlon = ee.Image.pixelLonLat().addBands(ndti)
latlon = latlon.reduceRegion(
    reducer=ee.Reducer.toList(),
    geometry=region,
    scale=100)
data_ndti = np.array((ee.Array(latlon.get('ndti')).getInfo()))

#ndsi data
latlon = ee.Image.pixelLonLat().addBands(ndsi)
latlon = latlon.reduceRegion(
    reducer=ee.Reducer.toList(),
    geometry=region,
    scale=100)
data_ndsi = np.array((ee.Array(latlon.get('ndsi')).getInfo()))

```
